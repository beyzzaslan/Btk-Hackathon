# TrustLens AI - Ekip Senkronizasyon ve Durum Raporu 🚀

Bu belge, projenin Backend geliştiricisi tarafından hazırlanmıştır. Şu ana kadar yapılanları ve Frontend ile AI (Yapay Zeka) sorumlusu takım arkadaşlarının bundan sonra atması gereken adımları özetler.

## 1. Backend Tarafında Neler Yaptım? (Mevcut Durum)
- **FastAPI Altyapısı:** Hızlı ve asenkron çalışan bir API (`POST /scan`) kuruldu.
- **Scraping Servisi:** Trendyol gibi sitelerden ürün adı, fiyat ve kampanya etiketlerini başarıyla çekiyoruz.
- **Veritabanı Yok (No DB):** Hackathon'un ruhuna uygun olarak sistem anlık (real-time) çalışıyor. Tarama geçmişi veya kullanıcı logini gibi özellikler olmadığı için **bilerek herhangi bir veritabanı (DB) veya tablo YAPILMADI**. Her şey anlık analiz edilip sonuç dönüyor.
- **Frontend Uyumlu Çıktı:** Tasarımcı arkadaşımın hazırladığı ekranlara *birebir* uyumlu olacak şekilde 5 farklı skor ve analiz kartı üreten detaylı bir JSON iskeleti hazırlandı.

## 2. Projeyi Bilgisayarınızda Nasıl Ayağa Kaldırırsınız? (Kurulum)
Projeyi kendi bilgisayarınızda test etmek için terminalde (veya CMD'de) sırasıyla şu adımları uygulayın:

```bash
# 1. Backend klasörüne girin
cd backend

# 2. Sanal ortam (virtual environment) oluşturun ve aktif edin
python -m venv venv
.\venv\Scripts\activate  # Windows için
# source venv/bin/activate # Mac/Linux için

# 3. Gerekli kütüphaneleri yükleyin
pip install -r requirements.txt

# 4. .env dosyası oluşturup içine GEMINI_API_KEY=senin_keyin şeklinde API anahtarını ekleyin

# 5. Sunucuyu başlatın
uvicorn app.main:app --reload
```
Sunucu başlayınca **http://127.0.0.1:8000/docs** adresinden Swagger arayüzüne girip test yapabilirsiniz.

## 3. Frontend Geliştiricisi Ne Yapmalı?
Tasarım çok başarılı! Hazırladığın arayüzü doğrudan besleyecek JSON çıktısını API'den şu şekilde alacaksın:

```json
{
  "url": "https://www.trendyol.com/...",
  "product_name": "Mad Parfüm...",
  "product_price": "599,99 TL",
  "trust_score": 72,
  "trust_label": "Dikkatli Ol",
  "score_review_quality": 48,
  "score_discount_reality": 60,
  "score_manipulation": 55,
  "score_domain": 30,
  "analysis_review": {
    "status": "Şüpheli",
    "description": "48 yorumdan 31'i şüpheli bulundu.",
    "badges": ["Aynı gün 15 yorum", "Aşırı jenerik dil", "%97 beş yıldız"]
  },
  "analysis_discount": {
    "status": "Yüksek risk",
    "description": "4.999₺ -> 299₺ iddiası doğrulanamadı.",
    "badges": ["Sayaç sıfırlanıyor", "Referans fiyat yok"]
  },
  "analysis_manipulation": {
    "status": "2 teknik",
    "description": "Yapay aciliyet ve kayıp kaçınımı dili tespit edildi.",
    "badges": ["Son 2 ürün kaldı!", "Fırsatı kaçırma"]
  },
  "analysis_domain": {
    "status": "Dikkat",
    "description": "Domain 14 gün önce açılmış...",
    "badges": ["14 günlük domain", "SSL ✓"]
  },
  "analysis_content": {
    "status": "Normal",
    "description": "Fiyat ve ürün açıklaması birbiriyle tutarlı...",
    "badges": []
  }
}
```

## 4. Yapay Zeka (AI) Geliştiricisi Ne Yapmalı?
Sistem şu an `Google Gemini-Flash` modeli ile entegre. Ancak AI tarafında senin dokunuşuna ihtiyaç var.

- **Yorumları Çekme Sıkıntısı:** Bilmen gereken en önemli şey; Trendyol'un agresif bot koruması (Akamai) nedeniyle şu an organik kullanıcı yorumlarını stabil bir şekilde **çekemiyoruz**. Yorumlar sayfaya JavaScript ile gizli yüklendiğinden bota düşmüyor.
- **Görev 1:** `backend/app/services/gemini_service.py` dosyasındaki *Prompt'u (Komutu)* detaylandırıp eğitmelisin. Sınırlı metinler üzerinden dark pattern ve sahte indirim yakalama yeteneğini artırmalısın.
- **Görev 2:** Frontend'deki `badges` (rozetler) ve `status` mantığının dinamik olarak düzgün üretilmesini sağlamalısın. Prompt içerisindeki JSON kurallarını sıkılaştır.
